```python
weather_df = pd.read_csv('../Outputs/All_Cities.csv')

del weather_df['Unnamed: 0']

weather_df
```

```python
location = weather_df[['Latitude', 'Longitude']].astype(float)

humid = weather_df[['Humidity (%)']].astype(float)

humid
```

```python
maps = gmaps.figure()

layer = gmaps.heatmap_layer(
    location, 
    weights = humid, 
    dissipating=False, 
    max_intensity=10, 
    point_radius=1)

maps.add_layer(layer)

maps
```

```python
vaca_df = weather_df.copy()
vaca_df = vaca_df.loc[vaca_df['Temperature (F)'] > 70]
vaca_df = vaca_df.loc[vaca_df['Temperature (F)'] < 80]
vaca_df = vaca_df.loc[vaca_df['Wind Speed (mph)'] < 10]
vaca_df = vaca_df.loc[vaca_df['Cloudiness (%)'] == 0]

vaca_df
```

```python
hotel_df = vaca_df.copy()
hotel_df = hotel_df.reset_index()
hotel_df['Hotel Name'] = ''
hotels = []

hotel_df
```

```python
print('Searching')
print('-------------------------')

for i, row in hotel_df.iterrows():
    
    target_coordinates = str(row['Latitude']) + ',' +  str(row['Longitude'])    
    target_search = "Hotel"
    target_radius = 5000
    target_type = "lodging"
    
    params = {
        "location": target_coordinates,
        "keyword": target_search,
        "radius": target_radius,
        "type": target_type,
        "key": g_key
    }
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    response = requests.get(base_url, params=params)
    hold = response.json()
    
    try:
        hotels.append(hold['results'][0]['name'])
        print(hold['results'][0]['name'])
    
    except (KeyError, IndexError):
        print("Hotel not found")
        hotels.append('N/A')

print('=========================')
print('Search complete')
```

```python
hotel_hold = pd.DataFrame(hotels)
hotel_hold = hotel_hold.drop(hotel_hold.index[12 : 36])

hotel_hold
```

```python
frames = [hotel_df, hotel_hold]

comb_df = pd.concat(frames, axis=1)

comb_df
```

```python
comb_df = comb_df.rename(columns={comb_df.columns[10] : 'Hotel Names'})
del comb_df['index']
del comb_df['Hotel Name']
comb_df = comb_df[comb_df['Hotel Names'] != 'N/A'].reset_index(drop=True)

comb_df
```
